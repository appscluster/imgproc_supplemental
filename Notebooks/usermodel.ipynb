Let's make a canvas of triangles and circles of varying size

For the tutorial, we generate a contrived canvas, but in most use cases, the canvas would be generated directly from an image via Canvas.from_labels().

```python
c = Canvas.random_circles() + Canvas.random_triangles()
c.show();
```

The next step is to define classes that inherit from the **Descriptor** class.  Each descriptor requries the following:

- **classifier**: str
   - simple logic function like (radius < 50). 
    
- **color**: str or None

- **alias**: str or None
   - If None, Descriptor is reference via the class name.

```python
from pyparty.scripts import Descriptor

class Small(Descriptor):
    classifier = '(d < 30)'
    color = 'r'

class SmallCircles(Descriptor):
    classifier = '(d < 30) & (eccentricity == 0.0)'
    color = 'orange'        
    alias = 'little circles'
        
class Large(Descriptor):
    classifier = '(d > 30)'
    color = 'green'

class LargeTriangles(Descriptor):
    classifier = '(d > 30) & (eccentricity > 0.01)'
    color = None
    
four_descriptors = [Small(), SmallCircles(), Large(), LargeTriangles()]
```

Classifiers can combine an abitrary sequence of logic functions joined by [& , |] operators, denoting [and, or] respectively.  The variable must be a valid pyparty descriptor (area, perim etc...) OR a custom function.  The following shortcuts and custom functions are implicitly builting:

- d --> equivalent_diameter
- r --> radius
- circularity 

Where the definition of radius looks something like:

    def radius(obj):

        return getattr(obj, 'equivalent_diameter')/2.0
        
Some caveats on the logic functions/classifier syntax:

- All logic functions must be in paranthesis (eg **d < 50** vs. **(d < 50)** )
- Complex/nested expression are forbidden for now eg:
    - ((d<50) & (circularity == 0.0)) | (d > 100)

Pyparty should catch these cases, but it is using a hacky, custom parser, so it's best not to tempt it.

The next step is to simply create a **Model** object with these Descriptors.  *Model* is a light container that essentially parses a Canvas based on the user-defined Descriptors, and interfaces to a *MultiCanvas*.

```python
from pyparty.scripts import Model

model = Model(*four_descriptors)
model
```

We use the **model.to_multicanvas**(canvas, *kwargs*) method to return a MultiCanvas.  The keyword options are:

- **sequential** (bool False):
    - If True, particles are removed from canvas after consumed by a descriptor.  Therefore, the same particle will not be found in multiple descriptors.
- **sans** (bool False):
    - If True, the condition in the Descriptor is reversed.  EG if descriptor was area < 100, all of area > 100 would be returned.  
    - In the example below, notice that SmallCircles (little circles) is empty because all of these would have been consumed in the more general*Small* descriptor.
- **modnames** (bool True):
    - If *sans*, "sans" is appended in front of particle alias to indicate the null condition.
    
These cases are best distinguished in the example below.

```python
mc = model.to_multicanvas(c)
mc.show(names=True)

mc_sequential = model.to_multicanvas(c, sequential=True)
mc_sequential.show(names=True)

mc_null = model.to_multicanvas(c, sans=True)
mc_null.show(names=True);
```

Since *LargeTriangles* didn't have a color, one was auto-generated for it in the MultiCanvas.

```python
import numpy as np
import os.path as op

from skimage import io
from skimage.color import rgb2gray
from skimage.morphology import label
from pyparty import Canvas, splot, showim, MultiCanvas 
```

```python

def load_img(basename):
    """ Set relative path to digitization folder to use files/images therein"""
    relpath = op.join( '../images/Test_Data/Ilastik_Analysis', basename)
    return io.imread(relpath)

def load_testimg(basename):
    """ Set relative path to digitization folder to use files/images therein"""
    relpath = op.join( '../images/Test_Data//Version1', basename)
    return io.imread(relpath)
```

```python
shape = load_img('SEM_test_noise_gray.png')
colorshape = load_img('SEM_test_noise.png')

shape_2 = load_img('SEM_test_nosmooth_gray.png')
colorshape_2 = load_img('SEM_test_nosmooth.png')

shape_3 = load_img('SEM_test_smooth_gray.png')
colorshape_3 = load_img('SEM_test_smooth.png')


noise = load_testimg('SEM_test_noise.png')
nosmooth = load_testimg('SEM_test_nosmooth.png')
smooth = load_testimg('SEM_test_smooth.png')

NAMES = ('singles', 'dimers', 'trimers', 'clusters')
```

## test_noise.png

```python
ax1, ax2, ax3 = splot(1,3, figsize=(12,6))

showim(noise, ax1,'gray', title='Raw image: test_noise.png');
showim(shape, ax2,'gray', title='Object classification gray');
showim(colorshape, ax3,'gray', title='Object classification in color');
```

```python
mc_noise = MultiCanvas.from_labeled(shape, *NAMES, ignore='black')
print mc_noise
```

```python
mc_noise.set_colors('r','g','y', 'magenta')
mc_noise.show(names=True)
mc_noise.mycolors
```

```python
ax1, ax2 = splot(1, 2, figsize=(10,5))

chartkwds = {'autopct':'percent', 'shadow':True}

mc_noise.pie(ax1, **chartkwds);
mc_noise.pie(ax2, attr='area', explode=(0,0,0,0.1), **chartkwds);
```

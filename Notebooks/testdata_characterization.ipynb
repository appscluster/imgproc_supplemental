```python
import numpy as np

from skimage import io
from skimage.color import rgb2gray
from skimage.morphology import label
from pyparty import Canvas, splot, showim 
```

```python
%cd ~/Desktop/images/test_image_results
filename1 = 'test_noise_shape_gray.png'
filename2 = 'test_noise_shape.png'
shape = io.imread(filename1)
colorshape = io.imread(filename2)

filename3 = 'test_nosmooth_shape_gray.png'
filename4 = 'test_nosmooth_shape.png'
shape_2 = io.imread(filename3)
colorshape_2 = io.imread(filename4)

filename5 = 'test_smooth_shape_gray.png'
filename6 = 'test_smooth_shape.png'
shape_3 = io.imread(filename5)
colorshape_3 = io.imread(filename6)

%cd ~/Desktop/images
noise = io.imread('test_noise.png')
nosmooth = io.imread('test_nosmooth.png')
smooth = io.imread('test_smooth.png')

NAMES = ('singles', 'dimers', 'trimers', 'clusters')
```

## test_noise.png

```python
ax1, ax2, ax3 = splot(1,3, figsize=(12,6))

showim(noise, ax1,'gray', title='Raw image: test_noise.png');
showim(shape, ax2,'gray', title='Object classification gray');
showim(colorshape, ax3,'gray', title='Object classification in color');
```

```python
numpy.unique(shape)
```

```python
from pyparty import MultiCanvas

mc_noise = MultiCanvas.from_labeled(shape, *NAMES, ignore='black')
print mc_noise
```

```python
#axes = splot(2,2, figsize=(10,10))

#for idx, (name, image) inenumerate(mc_noise.items()):
#    mc_noise[idx].show(axes[idx], title=name)
 
mc_noise.set_colors('r','g','y', 'magenta')
mc_noise.show(names=True)
mc_noise.mycolors
```

```python
ax1, ax2 = splot(1, 2, figsize=(10,5))

chartkwds = {'autopct':'percent', 'shadow':True}

mc_noise.pie(ax1, **chartkwds);
mc_noise.pie(ax2, attr='area', explode=(0,0,0,0.1), **chartkwds);
```

## test_nosmooth.png

```python
ax1, ax2, ax3 = splot(1,3, figsize=(12,6))

showim(nosmooth, ax1,'gray', title='Raw image: test_nosmooth.png');
showim(shape_2, ax2,'gray', title='Object classification gray');
showim(colorshape_2, ax3,'gray', title='Object classification in color');
```

```python
from pyparty import MultiCanvas

mc_nosmooth = MultiCanvas.from_labeled(colorshape_2, *NAMES, ignore='white')
print mc_nosmooth

axes = splot(2,2, figsize=(10,10))

for idx, (name, image) in enumerate(mc_nosmooth.items()):
    mc_nosmooth[idx].show(axes[idx], title=name)
    

mc_nosmooth.set_colors('r','g','y', 'magenta')
mc_nosmooth.mycolors

ax1, ax2 = splot(1, 2, figsize=(10,5))

chartkwds = {'autopct':'percent', 'shadow':True}

mc_nosmooth.pie(ax1, **chartkwds);
mc_nosmooth.pie(ax2, attr='area', explode=(0,0,0,0.1), **chartkwds);
```

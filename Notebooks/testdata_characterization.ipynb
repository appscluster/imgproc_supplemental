A quick utility to return all three of these is **_shared.SEM_test_images()**

```python
from _shared import SEM_test_images, load_test_image

#CROP ALL IMAGES
AUTOCROP = (0,0,512,512)

basic, smooth, noisy = SEM_test_images(crop=AUTOCROP)
```

Object labels were assigned in **Ilastik**; we will import gray-labels and colored-labels

```python
import os.path as op
from skimage import io 

def load_ilastik(basename):
    """ Lead image from ILastik folder"""
    relpath = 'Test_Data/Ilastik_Analysis'
    return load_test_image(basename, relpath, crop=AUTOCROP)

shape = load_ilastik('SEM_test_noise_gray.png')
colorshape = load_ilastik('SEM_test_noise.png')

shape_2 = load_ilastik('SEM_test_nosmooth_gray.png')
colorshape_2 = load_ilastik('SEM_test_nosmooth.png')

shape_3 = load_ilastik('SEM_test_smooth_gray.png')
colorshape_3 = load_ilastik('SEM_test_smooth.png')
```

#### Step 2: Show test image vs. gray/color ilastik labels

```python
ax1, ax2, ax3 = splot(1,3, figsize=(12,6))

showim(noisy, ax1, title='Raw image: test_noise.png')
showim(shape, ax2, title='Object classification gray')
showim(colorshape, ax3, title='Object classification in color');
```

```python
NAMES = ('singles', 'dimers', 'trimers', 'clusters')
mc_noise = MultiCanvas.from_labeled(shape, *NAMES, ignore='black')
print mc_noise
```

```python
mc_noise.set_colors('r','g','y', 'magenta')
mc_noise.show(names=True)
mc_noise.mycolors
```

```python
ax1, ax2 = splot(1, 2, figsize=(10,5))

chartkwds = {'autopct':'percent', 'shadow':True}

mc_noise.pie(ax1, **chartkwds);
mc_noise.pie(ax2, attr='area', explode=(0,0,0,0.1), **chartkwds);
```

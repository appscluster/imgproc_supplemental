## Particle Counting using Scipy (ndimage label)

## Imports; plot settings; import image

Load a scanning electron microscope example image (1024x768) of 22-nm gold nanoparticles on a silanized glass
optical fiber.


```python
pylab.rcParams['figure.figsize'] = 8, 5.5
%config InlineBackend.figure_format = 'svg' #svg backend

import numpy as np
from skimage.io import imread
from scipy.ndimage import label # For counting particles

from utilities import zoom, crop, zoomplot
```

```python
image = imread('images/APRIL_29_f2_100000.tif')
imshow(image, plt.cm.gray);
```

### Crop a subsection of image; binarize

Cropping is quite straightforward.  We will choose a rectangle with coordinates:

    xi = 200, yi = 200
    xf = 600, yf = 500

This should yield a **300 X 400** pixel image, preserving the aspect ratio (although this is not required).

```python
subimage = crop(image, 200, 200, 600, 500)
zoomplot(image, 200, 200, 600, 500, plt.cm.gray)
```

We will manually binarize the image to 143 counts.  **Note that we do not invert the image before calling "label" functions**.

```python
binary = image > 143
imshow(binary, plt.cm.gray);
```

### Labeling/counting particles

Labeling is performed through scipy.ndimage.measurements.label(), with **structure set to None**. This functional is
not as user-friendly as ImageJ’s analyze particles routine, and I am not sure which structure parameter to use for
counting.

```python
particlearray, count = label(binary)
plt.hist(particlearray, bins=50)
plt.title("Histogram of particle counts (dominated by noise)")
plt.xlabel('Size (in area?)')
plt.ylabel('Counts')

print 'Found ', count, ' particles! Most of them are noise'
```

```python
plt.hist(particlearray, bins=50)
plt.title("Histogram of particle counts (limited axes)")
plt.xlabel('Size (in area?)')
plt.ylabel('Counts')
plt.xlim(500,6000)
plt.ylim(0, 5000);
```

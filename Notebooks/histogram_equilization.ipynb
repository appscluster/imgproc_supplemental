### Local Histogram Equalization

```python
import skimage
from pyparty.plots import multishow, grayhist

HISTKWARGS = {'cdf':'r', 'bins':256}
```

```python
# Load an example image
filename = '../images/LOWCONTRAST.tif'
IMAGE = skimage.io.imread(filename)
showim(IMAGE)
grayhist(IMAGE, xlim=(0,70), **HISTKWARGS);
```

```python
# Contrast stretching
p2 = np.percentile(IMAGE, 2)
p98 = np.percentile(IMAGE, 98)
img_rescale = skimage.exposure.rescale_intensity(IMAGE, in_range=(p2, p98))

# Equalization
img_eq = 255 * skimage.exposure.equalize_hist(IMAGE)

# Adaptive Equalization (outputs uint16 so take sqrt)
img_adapteq = np.sqrt(skimage.exposure.equalize_adapthist(IMAGE, clip_limit=0.01))

CORRECTED = [img_rescale, img_eq, img_adapteq]
NAMES = ['Stretching', 'Equilization', 'Adaptive Eq.']

multishow(CORRECTED, names=NAMES)
COLORS = ['green', 'gold', 'magenta']

for idx, ax in enumerate(splot(1,3)):
    ax.yaxis.set_visible(False)
    grayhist(CORRECTED[idx], ax, color=COLORS[idx], histtype='step', **HISTKWARGS);
```

```python
TOTAL_HIST = grayhist(IMAGE, bins=50, histtype='stepfilled', cdf='black')
for idx, img in enumerate(CORRECTED):
    color = COLORS[idx]
    grayhist(img, axes=TOTAL_HIST, color=color, histtype='stepfilled', bins=50, cdf=color);
```

Lets fit these bins with a function and fill under to make a more smooth looking version of this.

```python
from scipy.stats import norm 

XVALS = np.linspace(0, 255, 500)

def curve(img):
    mu, sigma = norm.fit(img)
    n, bins = np.histogram(img)
    return mlab.normpdf(XVALS, mu, sigma)    

multishow([IMAGE] + CORRECTED)

patches = []
AXOUT = splot(1,1)[0]

AXOUT.scatter(x, curve(img), color='black', lw=2, alpha=.3)
patches.append(AXOUT.fill_between(XVALS, curve(IMAGE), alpha=.8, color='black'))

for idx, img in enumerate(CORRECTED):
    color = COLORS[idx]
    AXOUT.scatter(x,curve(img), color=color, lw=2, alpha=.3)
    patches.append(AXOUT.fill_between(XVALS, curve(img), alpha=.8, color=color))
    
AXOUT.legend( ['Original'] + NAMES )
plt.xlabel('Pixel Intensity')
plt.ylabel("Number of Pixels")
plt.title("Histogram Equilization (3 methods)")
plt.tick_params(labelleft='off') #Turn off xlabel values
plt.xlim(0,255)
plt.ylim(0, 0.025);
```

**WIKI**: This method usually increases the global contrast of many images, especially when the usable data of the image is represented by close contrast values. Through this adjustment, the intensities can be better distributed on the histogram. This allows for areas of lower local contrast to gain a higher contrast. Histogram equalization accomplishes this by effectively spreading out the most frequent intensity values.

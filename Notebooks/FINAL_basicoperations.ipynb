### Build a scale to convert nm to pixels and vice versa using 3.27 nm/pixels

We use a custom **Converter** class to facilitate easy converstion between units of pixels and length (nanometers).

```python
nm2pix = Converter(unit='nanometers', scale=3.27)
nm2pix.show()
print '------------------'
#Use invert() method to construct inverse scale
pix2nm = nm2pix.invert()
pix2nm.show()
```

We can use convert and to test out some converstions

```python
nm2pix.convert(10)
```

The *show_convert* function is more explicit in regard to units, but still returns the same value. 

```python
nm2pix.show_convert(10)
```

 **Note:** that *show_convert()* only works for single values, while *convert()* works on single value or iterables

```python
valtype = lambda x: (x, type(x))

print 'converting float:',  valtype(nm2pix.convert( 10 ))
print 'converting set:',    valtype(nm2pix.convert( {10,20,30} ))
print 'converting list:',   valtype(nm2pix.convert( [10,20,30] ))
print 'converting tuple:',  valtype(nm2pix.convert( (10,20,30) ))
print 'converting ndarray:', valtype(nm2pix.convert( np.array((10,20,30)) ))
```

This can be especially useful for plotting an image with length units.  We know that our image has an 
extent from (0,1024) and (768, 0) in pixels, so we can convert it to nanometers in the **extent** keyword of imshow.

```python
fix, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# Notice order of extent is (0, ymax, xmax, 0)
pix_extent = (0,image.shape[1], image.shape[0], 0)
nm_extent = pix2nm.convert(pix_extent)

ax1.imshow(image, extent = pix_extent)
ax1.set_xlabel('pixels')
ax1.set_ylabel('pixels')
ax1.set_title('%s X %s image (pixels)' % (image.shape) )

ax2.imshow(image, extent = nm_extent)
ax2.set_xlabel('nm')
ax2.set_ylabel('nm') # Round to .1 nm
ax2.set_title('%0.1f X %0.1f image (nm)' %  tuple(pix2nm.convert(image.shape)));
```

Because plotting an image under the unit-scaled representation is so common, and the *extent* dimensions are a bit counter-intuitive, we developed a simple wrapper **scaleshow()** to do the above.

```python
scaleshow(image, converter=pix2nm);
```

At the moment, *scale_show()* in not compatable with zoom() and zoomshow(); however, one may crop an image and then use *scale_show()* to zoom into a region of interest (ROI).

```python
ZOOM = (0,0,80,60) # in pixels
cropped = crop(image, ZOOM)
scaleshow(cropped, converter=pix2nm);
```

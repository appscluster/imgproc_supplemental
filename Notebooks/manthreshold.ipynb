### Manual Thresholding

Note: Image quality is reduced in pdf; true images are much clearer in the notebook.

```python
plt.rcParams['figure.figsize'] = 12,8
import numpy as np
%config InlineBackend.figure_format = 'svg' #svg backend
from skimage.io import imread

image = imread('images/MARCH_NICE_F2_30000_3_low.tif')
imshow(image, plt.cm.gray)
```

This fiber we decided had a manual threshold of 112 count.  To manually set this in scikit, we first binarize the image using the boolean **>** operator, then *invert* the numpy array to get the usual thresholded image.

```python
manual_bw = image > 112
manual_bw = np.invert(manual_bw)
imshow(manual_bw, plt.cm.gray)
```

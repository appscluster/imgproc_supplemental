# Quantify Segmentation with Four Labels in Ilastik

### Objective

We quantify the false positive/false negative of segmentation with Ilastik. The results turn to be consistent with watershed result, which is aroun 2.7%. It indicates that the particles tend to be oversized due to the smooth function.

### Environment Setup

Configure notebook style (see NBCONFIG.ipynb), add imports and paths.  The **%run** magic used below <font color='red'>**requires IPython 2.0 or higher.**</font>

```python
%run NBCONFIG.ipynb

from skimage.color import rgb2gray
from skimage.morphology import label
from pyparty import Canvas, MultiCanvas 
from pyparty.plots import showim, multishow, splot
import os.path as op
from skimage import io 
from pyparty.tools.arraytools import segment_summary, differ
```

#### Step 1: Import Binary Test Image

Import the binary test image and convert it into **boolean** image.

```python
img = io.imread('SEM_test_binary.png') > 0
showim(img)
```

#### Step 2: Import Ilastik segmented images with four labels

Noisy, nosmooth and smooth images are segmented in Ilastik with the **SAME** classifier, which contains four labels. One label for the dimmest particle, one for the brightest particle, one for the brightest background, and one for the dimmest background.

```python
Ilastik_noise = io.imread('Noise_seg_4labels.png') > 0
Ilastik_nosmooth = io.imread('Nosmooth_seg_4labels.png') > 0
Ilastik_smooth = io.imread('Smooth_seg_4labels.png') > 0

ax1, ax2, ax3 = splot(1,3, figsize=(10,6))
showim(Ilastik_noise, ax1, title='Noise_seg_4labels.png')
showim(Ilastik_nosmooth, ax2, title='Nosmooth_seg_4labels.png')
showim(Ilastik_smooth, ax3, title='Smooth_seg_4labels.png')
```

#### Step 3: Quantify segmentation errors fp/fn 

Here is a **disadvantage** of Ilastik. The classifier was trained in SEM_test_noise and then applied on SEM_test_smooth and SEM_test_nosmooth. This explains why the segmentation results for nosmooth and smooth are worse.

```python
print 'SEM_test_noise.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(img,Ilastik_noise)
print 'SEM_test_nosmooth.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(img,Ilastik_nosmooth)
print 'SEM_test_smooth.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(img,Ilastik_smooth)

ax1, ax2, ax3 = splot(1,3, figsize=(10,6))
showim(differ(Ilastik_noise, img), ax1, title='Noise_differ')
showim(differ(Ilastik_nosmooth, img), ax2, title='Nosmooth_differ')
showim(differ(Ilastik_smooth, img), ax3, title='Smooth_differ')
```

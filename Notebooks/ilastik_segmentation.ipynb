# Quantify Segmentation with Four Labels in Ilastik

### Objective

To quantify the of segmentation results from Ilastik by looking at the false positives and false negatives between its output and the original, binary rasterized image.  The results turn to be consistent with watershed result, which is aroun 2.7%. It indicates that the particles tend to be oversized due to the smooth function.  Furthermore, we will see that because the classifier was trained on a noisy image, it actually performs poorly when applied to segment the non-noisy test image.

### Environment Setup

Configure notebook style (see NBCONFIG.ipynb), add imports and paths.  The **%run** magic used below <font color='red'>**requires IPython 2.0 or higher.**</font>

```python
%run NBCONFIG.ipynb

from _shared import load_test_image
from pyparty.plots import multishow
```

#### Step 1: Import Binary Test Image

Import the binary test image and convert it into **boolean** image.  This has very sharp edges, so expect the ilastik segmentation, which is based on the smoothed image, to slightly over-segment at particle boundaries.

```python
binary = load_test_image('SEM_test_binary.png', ).astype('bool')
showim(binary, nolabel=True, title='Binary raster test image');
```

#### Step 2: Import Ilastik segmented images with four labels

From the classifier trained on the noisy image, we import the resulting segmentation when applied to the other two test images (*smooth* and *raster*).  This classifier only had 4 labels in total; two added to the dimmest and brightest background regions, and likewise on the dimmest and brightest paticle (from quick inspection, maybe not the brightest in the entire image...)

```python
noisy = load_test_image('noisy_seg.png', relpath='Test_Data/Ilastik_segmentation').astype(bool)
raster = load_test_image('raster_seg.png', relpath='Test_Data/Ilastik_segmentation').astype(bool)
smooth = load_test_image('smooth_seg.png', relpath='Test_Data/Ilastik_segmentation').astype(bool)

#multishow(noisy, raster, smooth)
```

#### Step 3: Quantify segmentation errors fp/fn 

Here is a **disadvantage** of Ilastik. The classifier was trained in SEM_test_noise and then applied on SEM_test_smooth and SEM_test_nosmooth. This explains why the segmentation results for nosmooth and smooth are worse.

```python
from pyparty.tools.arraytools import segment_summary, differ

print 'noisy.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(binary,noisy)
print 'raster.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(binary,raster)
print 'smooth.png: FP=%s  FN=%s  Net error=%s\n' % segment_summary(binary,smooth)

ax1, ax2, ax3 = splot(1,3, figsize=(10,6))
showim(differ(noisy, binary), ax1, title='Noise_differ')
showim(differ(raster, binary), ax2, title='Nosmooth_differ')
showim(differ(smooth, binary), ax3, title='Smooth_differ');
```

We see indeed that particle edges are the primary area of discrepancy, and this is not surprising as the smoothing will tend to blue particle boundaries and cause over-segmentation; there's not way around it.  **The difference in error is almost double just by inconsistent features between images.**  Let's color the false positives and negatives to illustrate this better.

```python
from pyparty.utils import any2rgb, to_normrgb, zoom

def cbinary(boolarray, color = 'orange'):
    color = to_normrgb(color)
    return any2rgb(boolarray) * color

noisy_fp, noisy_fn = differ(noisy, binary), differ(binary, noisy)
noisy_fp, noisy_fn = cbinary(noisy_fp, (1,0,0)), cbinary(noisy_fn, (0,1,0))
zoom(noisy_fp + noisy_fn, coords=(500, 1000, 1000, 1500) );
plt.savefig('/home/glue/Desktop/noisy_ilastik.png')
```

```python
smooth_fp, smooth_fn = differ(smooth, binary), differ(binary, smooth)
smooth_fp, smooth_fn = cbinary(smooth_fp, (1,0,0)), cbinary(smooth_fn, (0,1,0))
zoom(smooth_fp + smooth_fn, coords=(500, 1000, 1000, 1500) );
plt.savefig('/home/glue/Desktop/smooth_ilastik.png')
```

#### Step 1: ::Import our images from a relative directory::

Let's show the cropped region of the noisy image used for analysis.  

```python
from pyparty import showim
RELPATH = 'fourier_filter_examples/'

BP_FILTER = load_test_image('bandpass_filter.png', relpath=RELPATH)
BP_THRESH = load_test_image('bandpass_thresh.png', relpath=RELPATH)
HP_FILTER = load_test_image('high_pass_filter.png', relpath=RELPATH)
IMG_BP = load_test_image('image_bandpass.png', relpath=RELPATH)
IMG_HP = load_test_image('image_highpass.png', relpath=RELPATH)
NOISY = load_test_image('noisy_cropped.png', relpath=RELPATH)

imshow(NOISY);
```

Before moving on, I highly recommend Paul Borke's [Image Filtering in the Frequency Domain](http://paulbourke.net/miscellaneous/imagefilter/) tutorial!  

#### Step 2: Generate a low-pass and band-pass filter.

We used ImageJ's filtering functions to quickly generate these filters based on some low and high values for the cutoff parameters presented in ImageJ's frequency domain filtering applet.

** The high-pass filter allows highly-varying frequency fluctuations through.**  

** The bandpass filter screens both the very high fluctuations and the very low fluctionations.**

The most effectively filter would screen high-fluctuations corresponding to noise and low-fluctuations corresponding to the slowly-changing background gradient.  Moderate brightness fluctuations like those occurring between particle boundary and background ideally would not be altered.

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(8,5))

ax1.imshow(BP_FILTER)
ax2.imshow(HP_FILTER)
ax1.set_title('Band Pass filter')
ax2.set_title('High Pass Filter');
```

#### Step 3: Apply them to the noisy image.

**Note**: I saved a smaller crop of the high-pass filtered image, so the modified image shape differs below.

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(8,5))

ax1.imshow(IMG_BP)
ax2.imshow(IMG_HP)
ax1.set_title('Band Pass')
ax2.set_title('High Pass');
```

The Band Pass effectively removed the noise and background, and for the most part, preserved the paticle boundaries.  We observe a slight *ringing* effect, which is common in frequency-domain filtering.  The high-pass filter did not filter the majority of noise; we expect this since noise variations are generally high-frequency (ie abrupt brightness changes over the extent of a few pixels).  

<font color='blue'>**By carefully tuning the band-pass filter, we were able to get rid of most of the noise better than any spatial filters.</font>**  However, this took many iterations of manual parameter tuning, and we are unaware of any routines to automatically optimize the parameters for frequency-domain optimization.  Thus, it didn't too well in our automatic workflow.

#### Step 4: Segment based on the filtered image

With the background and noise nicely removed, we were able to just find a simple pixel-cutoff threshold:

```python
ax = imshow(BP_THRESH)
ax.axes.set_title("Pixel cutoff segmentation");
```

Some of the particle boundaries are warped, but overall, it's a pretty good segmentation compared to the segmentations on the noisy image without filtering!  Frequency filtering is certainly a nice way to go,

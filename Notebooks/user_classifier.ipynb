#### Step 1: Label the test image via Canvas.from_labels()

We import the test image and save all particles on it in Canvas via Canvas.from_labels().

** UPDATE WITH NEW IMAGES/DATA **

```python
from skimage import io 
from pyparty.utils import crop

img = io.imread('SEM_test_binary.png')
#img = crop(img, (0,0,1000,1024)) #1000 cuts edges nicely
           
c = Canvas(background=img)
```

```python
clab = c.from_labels(exclude='black', pbinary=False, pmin=30, neighbors=4)
clab.show();
```

#### Step 2: Histogram of various descriptors

Let's take a look at the distribution of various shape (circularity, minor_axis_length) and size (equivalent_diameter, perimeter, area) descriptors.

```python
from pyparty.plots import splot

# Store various attribute arrays
areas = clab.area
diams = clab.equivalent_diameter
circ = clab.circularity
minor, major = clab.minor_axis_length, clab.major_axis_length
perim = clab.perimeter

# Plots
BINS = 37
ax1, ax2, ax3, ax4 = splot(2,2, figsize=(10,10))

ax1.hist(diams, BINS, color='blue')
ax1.set_title('Diameter Equiv')
ax1.set_xlabel('px')

ax2.hist(areas, BINS, color='r')
ax2.set_title('Area')
ax2.set_xlabel('px^2')

ax3.hist(circ, BINS, color='orange')
ax3.set_title('Circularity')

ax4.hist(minor, BINS, color=(0,1,0))
ax4.set_title('Minor Axis')
ax4.set_xlabel('px');
```

We see that shape descriptors have a very clear separation of single particles from others.  This is more pronounced in *circularity* than *minor axis*.  Let's explore the correlation between various combos of these parameters.  This will illustrate how various combinations of shape and size parameters give nature choices for selection of filter attributes. 

Let's define a basic scatter plot wrapper:

```python
def ax_scatter(ax, x, y, xlabel='', ylabel='', color='blue'):
    ax.scatter(x,y,color=color)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title('%s - %s' % (ylabel.title(), xlabel.title()))
    return ax

AXES = splot(4,2, figsize=(12,18))
ax_scatter(AXES[0], diams, circ, 'diameter', 'circularity')
ax_scatter(AXES[1], perim, circ, 'perim', 'circularity', 'r')
ax_scatter(AXES[2], areas, circ, 'area', 'circularity', 'g')
ax_scatter(AXES[3], areas, perim, 'area', 'perimeter', 'black')
ax_scatter(AXES[4], minor, circ, 'minor axis', 'circularity', 'orange')
ax_scatter(AXES[5], minor, diams, 'minor axis', 'diameter', 'pink')
ax_scatter(AXES[6], major, circ, 'major axis', 'circularity', 'gray')
ax_scatter(AXES[7], major, diams, 'major axis', 'diameter', 'brown');
```

#### Step 3: Define the classifier with four descriptors for four species.  These cuts were chosen to ensure no overlap (ie they divide the space eveny as shown below)

```python
from pyparty.scripts.objectmodeling import Descriptor

class Single(Descriptor):
    classifier = '(minor_axis_length < 40) & (circularity > 0.8)'
    color = 'red'

class Dimer(Descriptor):
    classifier = '(minor_axis_length < 40) & (circularity < 0.8)' #IS <= a bug, can't recall?
    color = 'green'        
    #alias = 'little circles'
        
class Trimer(Descriptor):
    classifier = '(minor_axis_length > 40) & (circularity > 0.5)' #IS <= a bug, can't recall?
    color = 'gold'

class Tetramer(Descriptor):
    classifier = '(minor_axis_length > 40) & (circularity < 0.5)' #IS <= a bug, can't recall?
    color = 'magenta'
    
four_descriptors = [ Single(), Dimer(), Trimer(), Tetramer() ]
```

```python
LINEKWDS = {'linestyle':'--', 'linewidth':1}
plt.scatter(minor, circ, color='orange')

plt.hlines(y=0.8, xmin=0, xmax=40, **LINEKWDS) # Single/dimer boundary
plt.vlines(x=40, ymin=0, ymax=1.6, **LINEKWDS) # Vertical boundary
plt.hlines(y=0.5, xmin=40, xmax=80, **LINEKWDS)
plt.title("Minor axis vs. Circularity")

plt.xlim(20, 70)
plt.ylim(0.2,1.1);
```

```python
from pyparty.scripts import Model

model = Model(*four_descriptors)
model
```

#### Step 4: Apply the classifier on the test image

```python
mc = model.to_multicanvas(clab)
print mc
mc.show(names=True, nolabel=True);
```

```python
mc.pie(autopct='both');
```

```python
ax = mc.scatter(annotate=True, attr1='minor_axis_length', attr2='circularity')
ax.set_title("Minor Axis vs. Circularity")
ax.set_xlim(20,75);
```

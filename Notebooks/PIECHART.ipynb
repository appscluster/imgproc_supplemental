# Pie chart of labeled image

### Objective

We generate a pie chart of a pre-colored image of particles.  This is very similar to the analysis performed in the [pyparty publication](http://openresearchsoftware.metajnl.com/article/view/jors.bh), and required pyparty as a dependency.  Didn't include this in the primary notebooks as it's not an image-processing notebook per-se.

### Environment Setup

Configure notebook style (see NBCONFIG.ipynb), add imports and paths.  The **%run** magic used below <font color='red'>**requires IPython 2.0 or higher.**</font>

```python
%run NBCONFIG.ipynb
%pylab inline
from __future__ import division
```

#### Step 1: Import labeled image.  This image already exists in pyparty's builtin data, so just use it from there

```python
from pyparty import MultiCanvas
from pyparty.plots import showim, splot
from pyparty.data import nanolabels

NANOLABELS = nanolabels()
showim(NANOLABELS, 'spectral', title='Labeled Nanoparticles');
```

```python
NAMES = ('clusters', 'trimers', 'dimers', 'singles')

mc = MultiCanvas.from_labeled(NANOLABELS,  *NAMES, ignore=0.0, pmin=10)
mc.set_colors('magenta', '#f9ea1a', (0,1,0), 'r')
mc.show(names = True, ncols=2, figsize=(7,5));
```

```python
mc
```

```python
ax = mc.pie(attr='area', labels=None, autopct=None)
ax.set_title('')

#Make plot perfect circle
plt.axis('equal');
```

```python
help(mc.pie)
```

### Get the area of all of the pixels in the objects, the pixels in the background and compute the surface coverage as 100.0 X (area of particles / total area)

```python
objarea = np.sum( (NANOLABELS > 0) )
objarea
```

```python
bgarea = np.sum( (NANOLABELS == 0) )
bgarea
```

```python
totalarea = objarea + bgarea
totalarea
```

```python
print 'coverage is %.2f%%' % (100*objarea/totalarea)
```
